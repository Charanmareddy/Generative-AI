### Fixed size chunking

```python
with open('data_file.txt','r') as file:
    data = file.read()
fixed_size = 10
chunks = []
for index in range(0,len(data),fixed_size):
    chunk = data[index:index+fixed_size]
    chunks.append(chunk)
print(chunks)

```

### Fixed size + overlapping chunking

```python
with open('data_file.txt','r') as file:
    data = file.read()
fixed_size = 20
overlap = 4
chunks = []
for index in range(0,len(data),fixed_size-overlap):
    chunk = data[index:index+fixed_size]
    chunks.append(chunk)
print(chunks)
```

### Sentence based Chunking

```python
# \n,.,!,?
import re
with open('data_file.txt','r') as file:
    data = file.read()
spliter = '[s]'
re.sub('[.]',spliter,data)
```
