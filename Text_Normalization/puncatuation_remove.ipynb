```python
import string
print(string.punctuation)
print(len(string.punctuation))
```

```python
with open('data.txt','r+') as fh:
    data = ''
    for char in fh.read():
        if char not in string.punctuation:
            data+=char
    with open('data.txt','w') as file:
        file.write(data)

# with open('data.txt','r+') as fh:
#     data = fh.read()
#     for char in data:
#         if char not in string.punctuation:
#             data=data.replace(char,'')
#     with open('data.txt','w') as file:
#         file.write(data)
```
