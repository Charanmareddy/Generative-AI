```python
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
# create a list of sample tokens and lemmatize them using the existing `lemmatizer`
tokens = ['houses','running', 'ran', 'cars', 'better', 'geese', 'wolves', 'is', 'was', 'children', 'studies']
```

```python
'''Parameters
word : str
The input word to lemmatize.

pos : str
The Part Of Speech tag. Valid options are "n" for nouns, "v" for verbs, "a" for adjectives, "r" for adverbs and "s" for satellite adjectives.
'''
```

```python
for word in tokens:
    print(word, lemmatizer.lemmatize(word,pos = 'v'))
```


by default it uses 'n' for nouns

* if we have more words its diffucult to identify nouns and verbs
* this we can handle by using spacy lemmatize

```python
def find_postag(tag):
    if tag.startwith('J'):
        return 'a'
    elif tag.startwith('N'):
        return 'n'
    elif tag.startwith('V'):
        return 'v'
    elif tag.startwith('R'):
        return 'r'
    else:
        return 'n'
```
