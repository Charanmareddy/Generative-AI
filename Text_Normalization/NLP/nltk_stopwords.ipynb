```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

* use word tokenize, read text file convert into tokens remove all stop words

```python
with open(r'C:\Users\chara\desktop\Gen_AI\Text_Normalization\data.txt','r') as file:
    data = file.read()
```

```python
data
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
tokens = word_tokenize(data)
sent_tokens = sent_tokenize(data)
```

removing stop words by using word_tokenize

```python
for word in tokens:
    if word not in stop_words:
        data = data.replace(word,'')

print(data)
```
