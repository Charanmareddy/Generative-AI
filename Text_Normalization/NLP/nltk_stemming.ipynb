```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
words = [
    "running","runs","ran","Easily","fairly","playing","played","plays",
    "connection","connections","connected","Connects","a","happiness","Happy","happily",
    "studies","studying","study","organization","organizational"
]

for word in words:
    print(f'{word} ----> {stemmer.stem(word)}')
```

- Its not converting all the words into its root form
- because stemming is an rule based it works based on some rules
- it will just removes the suffix
- stemming does 2 things
    - convert the all words into lower case
    - converts words into ists root form based on some rules bu removing suffix

```python
from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer('english',ignore_stopwords=True)
```

- Snowball stemmer supports multiple languages where as PorterStemmer supports only one language i.e english
- snowball stemmer keeps the words as it is where as porterstemmer will convert them into lower case

```python
snowball_words = [
    "running", "runs", "ran", "Easily", "fairly", "the","playing", "played",
    "plays", "connection", "connections", "Connected", "connects",
    "happiness", "happy", "happily", "studies", "a","studying", "study",
    "organization", "organizational"
]
```

```python
for word in snowball_words:
    print(f'{word} ---> {snowball_stemmer.stem(word)}')
```

```python
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
stop_words = stopwords.words('english')
# stop_words
snowball_stemmer = SnowballStemmer('english',ignore_stopwords=True)
snowball_stemmer.stopwords = stop_words
# snowball_stemmer.stopwords
for word in snowball_words:
    print(snowball_stemmer.stem(word))
```
