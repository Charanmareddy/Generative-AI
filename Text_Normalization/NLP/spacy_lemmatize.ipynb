```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
# in nltk we use wordnet lemmatizer, in spacy we have en_core_web_sm.
# here pos tagging is not required
# name entity recognition
```

nltk -> stemming -> word net lemmatizer(wordnet) -> we can access lemmatization in nltk
- tokenization
- pos tagging(not required, by default)
- lemmatization
- name entity recognition

in nltk, we use word_tokenize. -> from nltk.tokenize import word_tokenize, sent_tokenize
- it doesn't convert into tokens by default, we need to convert
- 

ADVANTAGE
- Converts data into tokens(By default it converts into tokens)
- convert to lower case
- uses pos tag itself
- Spacy is faster than nltk

```python
tokens = nlp('''Apply the Pipeline pipeline To Some Text.''')
print(type(tokens), tokens) # it does not returns the tokens
# internally it is converted into tokens
```

```python
# to check, it is converted to tokens

for word in tokens:
    print(word)
# it returns words, not characters. How?
```

```python
# it returns spacy.tokens.token.Token, so it is group of tokens
for word in tokens:
    print(type(word))
# tokenization is not required separately -> advantage 1
```

```python
# converts some words to lowercase
for token in tokens:
    print(token.lemma_)  # is not converting all tokens to lowercase
```

After performing all the text normalization, we apply lemmatization

```python
# returns True for stop words
for token in tokens:
    print(token.lemma_, token.is_stop)
```

```python
# collect all non stop words
lemmatize_tokens = [token.lemma_ for token in tokens if not token.is_stop]
# for token in tokens:
#     print(token.lemma_, token.is_stop)  # it returns True for stopwords
lemmatize_tokens # prints tokens without stop words
# no need to use lemmatize method
```

```python
# convert to token
# convert to lowercase
# lemmatization using pos tagging
```

```python
print(' '.join(lemmatize_tokens))
```
